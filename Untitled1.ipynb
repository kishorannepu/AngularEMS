```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks 
%matplotlib inline

data=pd.read_csv("/home/indriyn/Downloads/Pattu Section(Dec-3-2018).csv")
value=data['Value']
#! pip install scipy
peaks, _ = find_peaks(value,threshold=10)
plt.plot(value)
plt.plot(peaks,value[peaks],"x")
plt.show()

print(len(peaks))
print(peaks)
```
